```python
import os
import pandas as pd
import numpy as np

#print(os.getcwd())
#chargement du dataset
df= pd.read_csv("creditcard.csv")
print("Dimension du dataset: ", df.shape)
#info sur les colonnes, valeurs non nulles
df.info()
                
```

```python
#Vérification des valeurs manquantes
missing = df.isna().sum()
print(missing [missing > 0] if missing.sum() > 0 else "Aucune valeur manquante détectectée.")
```

```python
# Statistiques descriptives
df.describe().T
```

```python
#-- Analyse de la distribution des variables --
import matplotlib.pyplot as plt
import seaborn as sns

# Histogramme du montant des transactions
plt.figure(figsize=(12,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Distribution du montant des transactions")
plt.xlabel("Montant (€)")
plt.ylabel("Fréquence")
plt.show()

```

```python
#Ce graphique montre une forte asymétrie à droite (distribution très concentrée vers 0), ce qui signifie que la majorité des transactions ont des montants faibles. 
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['Time'], bins=50, kde=True, color='orange')
plt.title("Distribution temporelle des transactions (en secondes)")
plt.xlabel("Temps écoulé depuis la première transaction (s)")
plt.ylabel("Fréquence")
plt.show()<

```

```python
# Boxplot du montant (pour repérer les valeurs extrêmes)
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Amount'], color='salmon')
plt.title("Boxplot du montant des transactions")
plt.xlabel("Montant (€)")
plt.show()
```

```python
#Répartition des classes
classes = df['Class'].value_counts()
print(classes)

#Pourcentage de chaque classe
pourcentage = df['Class'].value_counts(normalize=True) * 100
print("\nRépartition en pourcentage :")
print(pourcentage)
    
#Visualisation du déséquilibre des classes
plt.figure(figsize=(6,4))
ax = sns.countplot(
    x='Class',
    data=df,
    hue='Class',
    palette={0: 'skyblue', 1: 'crimson'},
)

# Ajouter les valeurs exactes
for p in ax.patches:
    ax.annotate(f"{int(p.get_height()):,}",
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')

plt.title("Visualisation du déséquilibre entre les transactions")
plt.xlabel("Classe (0 = Normal, 1 = Fraude)")
plt.ylabel("Nombre de transactions")
plt.show()

```

```python
#Déséquilibre des classes (Fraude vs Transactions normales)

#Ce graphique met en évidence un fort déséquilibre entre les classes :

#- La classe `0` (transactions normales) représente 99.83% des données.
#- La classe `1` (transactions frauduleuses) ne représente que 0.17%
```

```python
#Exploration des relations entre variables
#Calcul de la matrice de corrélation
corr_matrix = df.corr()

# --- Visualisation de la matrice de corrélation ---
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Matrice de corrélation globale des variables")
plt.show()
```

```python
### Interprétation des résultats

- La matrice de corrélation montre que les variables sont globalement faiblement corrélées entre elles, ce qui est cohérent avec un jeu de données issu d’une transformation PCA.
- Les variables les plus liées à la fraude (classe 1) sont: V11, V4 et V2 avec une corrélation positive modérée.
- Les variables V14, V17 et V12 présentent une corrélation négative marquée, indiquant un comportement différent pour les transactions frauduleuses.
- Le boxplot du montant des transactions montre que les fraudes concernent souvent des montants concentrés dans une plage restreinte, tandis que les transactions normales ont une grande variabilité.
```

```python
#Corrélation avec la variable Class
plt.figure(figsize=(8,6))
corr_class = corr_matrix['Class'].drop('Class').sort_values(ascending=False)
sns.barplot(x=corr_class.values, y=corr_class.index, hue=corr_class.index, palette='coolwarm')
plt.title("Corrélation de chaque variable avec la variable cible (Class)")
plt.xlabel("Coefficient de corrélation")
plt.ylabel("Variables")
plt.show()

```

```python
#Comparaison du montant (Amount) selon la classe
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df, color='darkgrey')
plt.title("Distribution du montant selon la classe (Normal vs Fraude)")
plt.xlabel("Classe (0 = Normal, 1 = Fraude)")
plt.ylabel("Montant (€)")
plt.show()
```
